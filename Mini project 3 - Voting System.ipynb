```python
s="*"
d="-"
bjp_votes = 0
congress_votes = 0
aap_votes = 0
shivsena_votes = 0
nota = 0
invalid = 0
print(s*90)
print("\t\t\t\t\tVoting System")
print(s*90)
print("Press A to vote for Aam aadmi party.")
print("Press B to vote for Bharatiya Janta Party.")
print("Press C to vote for congress party.")
print("Press S to vote for Shivsena party.")
print("Press N to vote for Non Of The Above.")
print()
print(d*70)
while True:
    ch = str(input("\tEnter a vote : "))
    vote = ch.upper()
    if vote == 'C':
        congress_votes+=1
    elif vote == "B":
        bjp_votes+=1
    elif vote == "A":
        aap_votes += 1
    elif vote == "S":
        shivsena_votes += 1
    elif vote == "N":
        nota += 1
    elif vote == "Q":
        print()
        break
    else:
        print("\t\tThis Vote is Invalid.\n")
        invalid +=1
print(d*70)
        
print()
print("\tNumber of Bharatiya Janta Party votes :",bjp_votes)
print("\tNumber of Congress Party Votes:",congress_votes)
print("\tNumber of Aam Aadmi Party votes :",aap_votes)
print("\tNumber of Shivsena Party votes :",shivsena_votes)
print("\tNumber of the Non Of The Above Votes :",nota)
print("\tNumber of Invalid Votes Are :",invalid)
print()

l = []
l.extend([bjp_votes, congress_votes, aap_votes, shivsena_votes, nota])
l.sort()
max_votes = l[-1]

print("\t\t",s*60)
if max_votes == bjp_votes:
    print("\t\t\tBharatiya Janta Party wins the election......")
elif max_votes == congress_votes:
    print("\t\t\tCongress party wins the election......")
elif max_votes == aap_votes:
    print("\t\t\tAam Aadmi Party wins the election......")
elif max_votes == shivsena_votes:
    print("\t\t\tShivsena party wins the election......")
else:
    print("\t\t\tNOTA has the higest votes in the election, hence re-election will be held")
print("\t\t",s*60)
```
